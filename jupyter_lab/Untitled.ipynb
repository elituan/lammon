```python
class avl_Node(object):
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
        self.height = 1
         
class AVLTree(object):
         
    def insert_node(self, tree, value):
 
        if not tree:
            return avl_Node(value)
        elif value < tree.value:
            tree.left = self.insert_node(tree.left, value)
        else:
            tree.right = self.insert_node(tree.right, value)
 
        tree.height = 1 + max(self.avl_Height(tree.left),
                              self.avl_Height(tree.right))
 
        # Update the balance factor and balance the tree
        balanceFactor = self.avl_BalanceFactor(tree)
        if -balanceFactor > 1:
            if value < tree.left.value:
                return self.rightRotate(tree)
            else:
                tree.left = self.leftRotate(tree.left)
                return self.rightRotate(tree)
 
        if -balanceFactor < -1:
            if value > tree.right.value:
                return self.leftRotate(tree)
            else:
                tree.right = self.rightRotate(tree.right)
                return self.leftRotate(tree)
 
        return tree
    def avl_Height(self, tree):
        if not tree:
            return 0
        return tree.height
 
    # Get balance factore of the node
    def avl_BalanceFactor(self, tree):
        if not tree:
            return 0
        return  self.avl_Height(tree.right) - self.avl_Height(tree.left) 
     
    def avl_MinValue(self, tree):
        if tree is None or tree.left is None:
            return tree
        return self.avl_MinValue(tree.left)
 
    def preOrder(self, tree):
        if not tree:
            return
        print("{0} ".format(tree.value), end=" ")
        self.preOrder(tree.left)
        self.preOrder(tree.right)
         
    def leftRotate(self, b):
        a = b.right
        T2 = a.left
        a.left = b
        b.right = T2
        b.height = 1 + max(self.avl_Height(b.left),
                           self.avl_Height(b.right))
        a.height = 1 + max(self.avl_Height(a.left),
                           self.avl_Height(a.right))
        return a
 
     
    def rightRotate(self, b):
        a = b.left
        T3 = a.right
        a.right = b
        b.left = T3
        b.height = 1 + max(self.avl_Height(b.left),
                           self.avl_Height(b.right))
        a.height = 1 + max(self.avl_Height(a.left),
                           self.avl_Height(a.right))
        return a
            
```

```python
import sys
from heapq import heappop, heappush
 

# A class to store a heap node
# Each Node have attribute vertex and weight
class Node:
    def __init__(self, vertex, weight=0):
        self.vertex = vertex
        self.weight = weight
 
    # Override the __lt__() function to make `Node` class work with a min-heap
    def __lt__(self, other):
        return self.weight < other.weight
 
 
# A class to represent a graph object
class Graph:
    def __init__(self, edges, n):
        # allocate memory for the adjacency list
        # Although the nodes do not include 0, we assume that the 0 node exist (but now form edge to any nodes). 
        # Then, It will be easier to call the index
        self.adjList = [[] for _ in range(n)]
 
        # add edges to the directed graph
        for (source, dest, weight) in edges:
            self.adjList[source].append((dest, weight))
 
 
def get_route(prev, i, route):
    if i >= 0:
        get_route(prev, prev[i], route)
        route.append(i)
 
 
# Run Dijkstra’s algorithm on a given graph
def findShortestPaths(graph, source, n, dest):
 
    # create a min-heap and push source node having distance 0
    pq = []
    heappush(pq, Node(source))
 
    # set initial distance from the source to `v` as infinity
    dist = [sys.maxsize] * n
 
    # distance from the source to itself is zero
    dist[source] = 0
 
    # list to track vertices for which minimum cost is already found
    done = [False] * n
    done[source] = True
 
    # stores predecessor of a vertex (to a print path)
    prev = [-1] * n
 
    # run till min-heap is empty
    while pq:
 
        node = heappop(pq)      # Remove and return the best vertex
        u = node.vertex         # get the vertex number
 
        # do for each neighbor `v` of `u`
        for (v, weight) in graph.adjList[u]:
            if not done[v] and (dist[u] + weight) < dist[v]:        # Relaxation step
                dist[v] = dist[u] + weight
                prev[v] = u
                heappush(pq, Node(v, dist[v]))
 
        # mark vertex `u` as done so it will not get picked up again
        done[u] = True
 
    route = []
    for i in range(n):
        if i != source and dist[i] != sys.maxsize:
            get_route(prev, i, route)
            if i == dest:
                print(f'Path ({source} —> {i}): Minimum cost = {dist[i]}, Route = {route}')
            route.clear()
```
